```python
df1=df1.drop(['SizeRank','RegionName','RegionType'],axis=1)
```

```python
test_melt = df1.melt('StateName', var_name = 'date', value_name = 'zhvi')
test_melt['date'] = pd.to_datetime(test_melt['date'],yearfirst = True)
test_melt.info()
test_melt
```

```python
sns.set(rc={'figure.figsize':(20,15)})
sns.lineplot(x=test_melt['date'], y= test_melt['zhvi'], hue = test_melt['StateName'])
```
